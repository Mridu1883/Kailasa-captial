```python
import pandas as pd

# Load the uploaded dataset
df = pd.read_csv('NIFTY 50_daily_data.csv')

# Display the first few rows and column names for understanding
df.head(), df.columns
```

```python
pd.set_option('display.max_rows', None)
```

```python
import numpy as np

# Calculate RSI
def calculate_rsi(series, period=14):
    delta = series.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Calculate MACD and MACD Histogram
def calculate_macd(series, fast=12, slow=26, signal=9):
    exp1 = series.ewm(span=fast, adjust=False).mean()
    exp2 = series.ewm(span=slow, adjust=False).mean()
    macd = exp1 - exp2
    signal_line = macd.ewm(span=signal, adjust=False).mean()
    histogram = macd - signal_line
    return macd, signal_line, histogram

# Add indicators to the dataframe
df['rsi'] = calculate_rsi(df['close'])
df['macd'], df['signal_line'], df['macd_hist'] = calculate_macd(df['close'])

# Define MACD color categorization
df['macd_color'] = np.where(df['macd_hist'] > df['macd_hist'].shift(1), 'dark_green',
                            np.where(df['macd_hist'] < df['macd_hist'].shift(1), 'dark_red', 'neutral'))

df[['date', 'close', 'rsi', 'macd', 'signal_line', 'macd_hist', 'macd_color']].tail(10)
```

```python
initial_capital = 1e7  # ₹1 crore
position_size = 1  # One lot per trade (assuming constant exposure for simplicity)
entry_price = None
position = None  # 'long' or 'short'
entry_index = None

trades = []

# Iterate through data starting from 1 to allow for shift (yesterday's data)
for i in range(1, len(df) - 1):
    today = df.iloc[i]
    tomorrow = df.iloc[i + 1]

    # Avoid entering/exiting on the same candle
    if position is None:
        # LONG CONDITION
        if today['rsi'] > 55 and today['macd_color'] == 'dark_green':
            if tomorrow['open'] > today['high']:
                entry_price = tomorrow['open']
                position = 'long'
                entry_index = i + 1
            elif tomorrow['high'] > today['high']:
                entry_price = today['high']
                position = 'long'
                entry_index = i + 1
        # SHORT CONDITION
        elif today['rsi'] < 45 and today['macd_color'] == 'dark_red':
            if tomorrow['open'] < today['low']:
                entry_price = tomorrow['open']
                position = 'short'
                entry_index = i + 1
            elif tomorrow['low'] < today['low']:
                entry_price = today['low']
                position = 'short'
                entry_index = i + 1

    # Handle exit if in position
    elif position is not None:
        candle = df.iloc[i]
        high, low = candle['high'], candle['low']
        exit_price = None
        exit_reason = None

        if position == 'long':
            if high - entry_price >= 150:
                exit_price = entry_price + 150
                exit_reason = 'TP'
            elif entry_price - low >= 75:
                exit_price = entry_price - 75
                exit_reason = 'SL'
        elif position == 'short':
            if entry_price - low >= 150:
                exit_price = entry_price - 150
                exit_reason = 'TP'
            elif high - entry_price >= 75:
                exit_price = entry_price + 75
                exit_reason = 'SL'

        if exit_price:
            trades.append({
                'entry_date': df.iloc[entry_index]['date'],
                'exit_date': candle['date'],
                'position': position,
                'entry_price': entry_price,
                'exit_price': exit_price,
                'exit_reason': exit_reason,
                'profit': (exit_price - entry_price) * (1 if position == 'long' else -1)
            })
            position = None
            entry_price = None
            entry_index = None

# Create trades DataFrame
trades_df = pd.DataFrame(trades)
trades_df.tail()
```

```python
from math import sqrt

# Compute results
total_profit = trades_df['profit'].sum()
num_trades = len(trades_df)
win_trades = trades_df[trades_df['profit'] > 0]
loss_trades = trades_df[trades_df['profit'] < 0]
success_percentage = len(win_trades) / num_trades * 100

# Sharpe ratio assuming risk-free rate = 0 and daily frequency
returns = trades_df['profit']
sharpe_ratio = returns.mean() / returns.std() * sqrt(252) if returns.std() != 0 else 0

{
    'Total Trades': num_trades,
    'Total Profit (₹)': total_profit,
    'Success Rate (%)': round(success_percentage, 2),
    'Sharpe Ratio': round(sharpe_ratio, 2)
}
```
